## Analisando a coluna is_fraud

```python
df.describe()
```

```python
df["trans_hours"]= pd.to_datetime(df['trans_time'], format='%H:%M:%S').dt.hour
```

```python
df.head()
```

```python
df.info()
```

```python
#Drop da coluna para limpar o dataset, como isso seria um dado irrelevante
df= df.drop(columns=["Unnamed: 0"])
```

```python
df.head()
```

```python
df_fraud = df[df["is_fraud"]==1]

if not df_fraud.empty:
    print("Fraude: Horario: Dia Fraudado:")
    print(df_fraud[["trans_hours","trans_date", "amt"]])
else:
    print("sem fraudes")
```

```python
df_fraud
```

```python
df_fraud["trans_hours"].value_counts(normalize=True)*100
```

```python
df_fraud["amt"].mean()
```

```python
df_fraud["amt"].describe()
```

```python
#Criação da lista com valores fraudados que aparecem nos Quartis da descrição
#Primeiro criamos uma lista com os valores que aparecem na descrição
faixa = [0, 50, 100, 200, 500, 1000, 10000]
#Regra(etiqueta), para organizar onde começa e onde termina a identificação das fraudes
faixa_percorre = ["0 a 50", "50 a 100", "100 a 200", "200 a 500", "500 a 1000", "A cima de 1000"]
#Organização de valores continuos para a identificação dos valores que mais apresentam fraudes
fraud_padrao = pd.cut(df_fraud["amt"],bins= faixa, labels= faixa_percorre).value_counts()
print("valores com mais fraudes")
print(fraud_padrao)
```

```python
fraud_padrao.plot(kind="bar", color="red", edgecolor="black")

# Textos do gráfico
plt.title("Quantidade de Fraudes por Faixa de Valor")
plt.xlabel("Faixas de Valor (R$)")
plt.ylabel("Número de Fraudes")

# Gira os nomes das faixas para não ficarem encavalados
plt.xticks(rotation=45) 
plt.savefig('grafico_fraudes.png', bbox_inches='tight')
plt.show()
```
